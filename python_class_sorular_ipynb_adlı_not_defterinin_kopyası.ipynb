### Soru 2 — Metotlar
Bir `Dikdortgen` class'ı yaz:
- `en` ve `boy` alsın
- `alan()` → en × boy döndürsün
- `cevre()` → 2 × (en + boy) döndürsün
- `kare_mi()` → en == boy ise `True` döndürsün

```python
# Cevabını buraya yaz
class Dikdortgen:
  def __init__(self,en,boy):
      self.en = en
      self.boy = boy
  def alan(self):
    return self.en*self.boy

  def cevre(self):
    return 2*(self.en+self.boy)

  def kare_mi(self):
    return self.en == self.boy


dd1=Dikdortgen(12, 3)
print(dd1.alan())
print(dd1.cevre())
print(dd1.kare_mi())




```

```python
# TEST
d1 = Dikdortgen(5, 3)
d2 = Dikdortgen(4, 4)
assert d1.alan() == 15
assert d1.cevre() == 16
assert d1.kare_mi() == False
assert d2.kare_mi() == True
print("✅ Soru 2 doğru!")
```

### Soru 5 — Kalıtım
`Hayvan` class'ından türeyen `Kopek` ve `Kedi` class'ları yaz:
- `Hayvan`: `isim` alsın, `ses_cikar()` → `"..."` döndürsün
- `Kopek`: `ses_cikar()` → `"Hav!"` döndürsün, `getir()` → `"[isim] topu getirdi!"` yazdırsın
- `Kedi`: `ses_cikar()` → `"Miyav!"` döndürsün

```python
class Hayvan:
    def __init__(self, isim):
        self.isim = isim

    def ses_cikar(self):
        return "..."

class Kopek(Hayvan):
    def ses_cikar(self):
        return "Hav!"

    def getir(self):
        print(f"{self.isim} topu getirdi!")

class Kedi(Hayvan):
    def ses_cikar(self):
        return "Miyav!"
```

```python
# TEST
k = Kopek("Karabaş")
c = Kedi("Pamuk")
assert k.isim == "Karabaş"
assert k.ses_cikar() == "Hav!"
assert c.ses_cikar() == "Miyav!"
assert isinstance(k, Hayvan)
assert isinstance(c, Hayvan)
k.getir()  # Karabaş topu getirdi!
print("✅ Soru 5 doğru!")
```

### Soru 6 — Kapsülleme ve @property
Bir `Sicaklik` class'ı yaz:
- `_derece` ile saklansın
- `@property` ile `derece` okunabilsin
- `@derece.setter` ile atanabilsin ama `-273`'ün altına düşürülemesin (mutlak sıfır), düşürülmeye çalışılınca `ValueError` fırlatsın
- `@property` ile `fahrenheit` hesaplansın → `(derece × 9/5) + 32`

```python
from IPython.core import debugger
class Sicaklik:
    def __init__(self, derece):
        self.derece = derece

    @property
    def derece(self):
        return self._derece
    @derece.setter
    def derece(self,derece):
      if deger < -273:
        ValueError
      self._derece = derece

    @property
    def fahrenheit(self):
      return (self._derece * 9/5) + 32


```

```python
# TEST
s = Sicaklik(100)
assert s.derece == 100
assert s.fahrenheit == 212.0
s.derece = 0
assert s.fahrenheit == 32.0
try:
    s.derece = -300
    assert False, "ValueError bekleniyor"
except ValueError:
    pass
print("✅ Soru 6 doğru!")
```

### Soru 7 — Dunder metotlar
Bir `Para` class'ı yaz:
- `miktar` ve `para_birimi` alsın (varsayılan: `"TL"`)
- `__str__`  → `"100 TL"` formatında
- `__add__`  → iki Para toplanabilsin (aynı para birimi kontrolü yap, farklıysa `ValueError`)
- `__eq__`   → miktar ve para birimi eşitse `True`
- `__lt__`   → miktar karşılaştırması

```python
class Para:
    def __init__(self, miktar, para_birimi="TL"):
        self.miktar = miktar
        self.para_birimi = para_birimi

    def __str__(self):
        return f"{self.miktar} {self.para_birimi}"

    def __add__(self, farkli):
        if self.para_birimi != farkli.para_birimi:
            raise ValueError("Sikinti: Para birimleri uyuşmuyor!")
        # Tuple yerine yeni bir Para nesnesi döndürüyoruz:
        return Para(self.miktar + farkli.miktar, self.para_birimi)

    def __eq__(self, farkli):
        return self.miktar == farkli.miktar and self.para_birimi == farkli.para_birimi

    def __lt__(self, farkli):
        if self.para_birimi != farkli.para_birimi:
            print("Uyarı: Farklı birimler karşılaştırılıyor!")
        return self.miktar < farkli.miktar
```

```python
# TEST
p1 = Para(100, "TL")
p2 = Para(50, "TL")
p3 = Para(100, "TL")
p4 = Para(50, "USD")
assert str(p1) == "100 TL"
assert (p1 + p2).miktar == 150
assert p1 == p3
assert p2 < p1
try:
    p1 + p4
    assert False, "ValueError bekleniyor"
except ValueError:
    pass
print("✅ Soru 7 doğru!")
```

### Soru 8 — @classmethod ve @staticmethod
Bir `Dikdortgen` class'ı yaz:
- `en` ve `boy` alsın
- `@classmethod kare_olustur(cls, kenar)` → kare döndürsün (en == boy == kenar)
- `@staticmethod gecerli_mi(en, boy)` → ikisi de 0'dan büyükse `True` döndürsün

```python
class Dikdortgen:
    def __init__(self, en, boy):
        self.en = en
        self.boy = boy

    @classmethod
    def kare_olustur(cls, kenar):
        """Sınıfın kendisini (cls) kullanarak bir kare nesnesi üretir."""
        return cls(kenar, kenar)

    @staticmethod
    def gecerli_mi(en, boy):
        """Nesneye veya sınıfa ait verilere ihtiyaç duymaz, genel bir kontrol yapar."""
        return en > 0 and boy > 0


boyut_kontrol = Dikdortgen.gecerli_mi(10, -5)
print(f"Boyutlar geçerli mi?: {boyut_kontrol}")

kare = Dikdortgen.kare_olustur(10)
print(f"Kare Alanı: {kare.en * kare.boy}")
```

```python
# TEST
d = Dikdortgen(5, 3)
kare = Dikdortgen.kare_olustur(4)
assert kare.en == 4
assert kare.boy == 4
assert Dikdortgen.gecerli_mi(5, 3) == True
assert Dikdortgen.gecerli_mi(-1, 3) == False
assert Dikdortgen.gecerli_mi(0, 3) == False
print("✅ Soru 8 doğru!")
```

### Soru 9 — Iterator class
Bir `AralıkIterator` class'ı yaz:
- `baslangic`, `bitis`, `adim` alsın
- `for` döngüsünde kullanılabilsin
- Python'un `range()` gibi çalışsın ama class olarak
- `__len__` da ekle — kaç eleman üreteceğini döndürsün

```python
class AralikIterator:
    def __init__(self, baslangic, bitis, adim=1):
        self.baslangic = baslangic
        self.bitis = bitis
        self.adim = adim
        self.mevcut = baslangic

    def __iter__(self):
        return self

    def __next__(self):
        if (self.adim > 0 and self.mevcut >= self.bitis) or \
           (self.adim < 0 and self.mevcut <= self.bitis):
            raise StopIteration

        sonuc = self.mevcut
        self.mevcut += self.adim
        return sonuc

    def __len__(self):
        if (self.adim > 0 and self.baslangic >= self.bitis) or \
           (self.adim < 0 and self.baslangic <= self.bitis):
            return 0

        return abs((self.bitis - self.baslangic + (self.adim - (1 if self.adim > 0 else -1))) // self.adim)

# --- Test Kodu ---
sayilar = AralikIterator(0, 10, 2)

print(f"Toplam üretilecek eleman sayısı: {len(sayilar)}")

print("Sayılar döngüsü başlıyor:")
for sayi in sayilar:
    print(sayi)
```

```python
# TEST
a = AralikIterator(0, 10, 2)
assert list(a) == [0, 2, 4, 6, 8]
assert len(a) == 5
b = AralikIterator(1, 6, 1)
assert list(b) == [1, 2, 3, 4, 5]
# iki kez kullanılabilmeli
assert list(a) == [0, 2, 4, 6, 8]
print("✅ Soru 9 doğru!")
```

### Soru 10 — Mixin
Aşağıdaki iki Mixin'i yaz ve `Calisan` class'ına ekle:
- `KarsilastirmaMixin`: `__eq__` ve `__lt__` eklesin — karşılaştırma `maas` değerine göre yapılsın
- `YazdirmaMixin`: `yazdir()` metodu tüm instance değişkenlerini `anahtar: değer` formatında yazdırsın
- `Calisan`: `isim`, `departman`, `maas` alsın

```python
# Cevabını buraya yaz
class KarsilastirmaMixin:
    # __eq__: Eşit mi? (Equal)
    def __eq__(self, diger):
        if not isinstance(diger, Calisan):
            return False
        return self.maas == diger.maas

    # __lt__: Küçük mü? (Less Than)
    def __lt__(self, diger):
        if not isinstance(diger, Calisan):
            return NotImplemented
        return self.maas < diger.maas

class YazdirmaMixin:
    def yazdir(self):
        bilgiler = vars(self)
        print("\n--- Çalışan Bilgileri ---")
        for anahtar, deger in bilgiler.items():
            print(f"{anahtar.capitalize()}: {deger}")

class Calisan(KarsilastirmaMixin, YazdirmaMixin):
    def __init__(self, isim, departman, maas):
        self.isim = isim
        self.departman = departman
        self.maas = maas
```

```python
# TEST
c1 = Calisan("Ali", "Yazılım", 50000)
c2 = Calisan("Ayşe", "Tasarım", 60000)
c3 = Calisan("Ali", "Yazılım", 50000)
assert c1 < c2
assert c2 > c1
assert c1 == c3
c1.yazdir()
# isim: Ali
# departman: Yazılım
# maas: 50000
print("✅ Soru 10 doğru!")
```

### Soru 11 — Hepsini Birleştir (Proje)
Bir `Kutuphane` sistemi yaz:

**`Kitap` class'ı:**
- `isbn`, `ad`, `yazar`, `mevcut` (varsayılan `True`) alsın
- `__str__` → `"[ISBN] Ad — Yazar"`
- `__repr__` → `"Kitap(isbn, ad, yazar)"`

**`Kutuphane` class'ı:**
- `kitaplar` listesi tut
- `kitap_ekle(kitap)` → listeye ekle
- `__len__` → toplam kitap sayısı
- `__contains__` → ISBN ile kitap var mı kontrolü (`"978..." in kutuphane`)
- `kitap_bul(isbn)` → ISBN ile kitap döndür, bulamazsa `None`
- `odunc_al(isbn)` → kitabı `mevcut=False` yap, zaten alınmışsa hata ver
- `iade_et(isbn)` → kitabı `mevcut=True` yap
- `mevcut_kitaplar()` → sadece mevcut kitapları listele

```python


class Kitap:
    def __init__(self, isbn, ad, yazar, mevcut=True):
        self.isbn = isbn
        self.ad = ad
        self.yazar = yazar
        self.mevcut = mevcut

    def __str__(self):
        # Kullanıcının göreceği format
        durum = "Mevcut" if self.mevcut else "Ödünç Verildi"
        return f"[{self.isbn}] {self.ad} — {self.yazar} ({durum})"

    def __repr__(self):
        # Geliştiricinin göreceği format
        return f"Kitap('{self.isbn}', '{self.ad}', '{self.yazar}')"


class Kutuphane:
    def __init__(self):
        self.kitaplar = []

    def kitap_ekle(self, kitap):
        if isinstance(kitap, Kitap):
            self.kitaplar.append(kitap)
            print(f"Sisteme eklendi: {kitap.ad}")
        else:
            print("Hata: Sadece Kitap nesnesi ekleyebilirsiniz.")

    def __len__(self):
        return len(self.kitaplar)

    def __contains__(self, isbn):
        # "isbn in kutuphane" kullanımını sağlar
        return any(k.isbn == isbn for k in self.kitaplar)

    def kitap_bul(self, isbn):
        for kitap in self.kitaplar:
            if kitap.isbn == isbn:
                return kitap
        return None

    def odunc_al(self, isbn):
        kitap = self.kitap_bul(isbn)
        if not kitap:
            print("Hata: Kitap bulunamadı.")
        elif not kitap.mevcut:
            print(f"Hata: '{kitap.ad}' zaten ödünç alınmış!")
        else:
            kitap.mevcut = False
            print(f"Başarılı: '{kitap.ad}' ödünç alındı.")

    def iade_et(self, isbn):
        kitap = self.kitap_bul(isbn)
        if kitap:
            kitap.mevcut = True
            print(f"Başarılı: '{kitap.ad}' iade edildi.")
        else:
            print("Hata: Geçersiz ISBN.")

    def mevcut_kitaplar(self):
        print("\n--- Mevcut Kitaplar Listesi ---")
        liste = [k for k in self.kitaplar if k.mevcut]
        for k in liste:
            print(k)
        if not liste:
            print("Şu an kütüphanede kitap yok.")
```

```python
# TEST
k1 = Kitap("978-1", "Suç ve Ceza", "Dostoyevski")
k2 = Kitap("978-2", "1984", "Orwell")
k3 = Kitap("978-3", "Dune", "Herbert")

kutuphane = Kutuphane()
kutuphane.kitap_ekle(k1)
kutuphane.kitap_ekle(k2)
kutuphane.kitap_ekle(k3)

assert len(kutuphane) == 3
assert "978-1" in kutuphane
assert "978-9" not in kutuphane

kutuphane.odunc_al("978-1")
assert k1.mevcut == False
assert len(kutuphane.mevcut_kitaplar()) == 2

kutuphane.iade_et("978-1")
assert k1.mevcut == True

try:
    kutuphane.odunc_al("978-1")
    kutuphane.odunc_al("978-1")  # zaten alınmış
    assert False
except Exception:
    pass

print("✅ Soru 11 doğru! Tebrikler!")
```
