### ✏️ Soru 24:
Bir alışveriş sepeti uygulaması yapın:
- `sepet = []` listesi oluşturun
- 3 ürün ekleyin (`.append()`)
- `for` döngüsü ile ürünleri numaralı yazdırın
- Kaç ürün olduğunu yazdırın

```python
sepet = []
sepet.append("ceket")
sepet.append("kemer")
sepet.append("ayakkabi")
sayac = 1
for i in sepet:
    print(sayac, i)
    sayac += 1

print("Toplam ürün:",len(sepet))

```

### ✏️ Soru 25:
Bir listede verilen notların ortalamasını hesaplayan `ortalama(notlar)` fonksiyonu yazın.
- `sum()` ve `len()` kullanabilirsiniz
- `[70, 85, 90, 60, 75]` listesiyle test edin

```python
# Buraya yazın:
def ortalama(notlar):
    toplam = sum(notlar)
    adet = len(notlar)
    return toplam/adet

test_notlari = [70, 85, 90, 60, 75]
print(ortalama(test_notlari))
# print(ortalama(test_notlari))  # 76.0 olmalı
```
