### 🔸 Soru 2:
Aşağıdaki ifadelerin Python'da geçerli olup olmadığını `try-except` kullanmadan, sadece `type()` ve mantık yürüterek tahmin edin, sonra çalıştırarak doğrulayın:
- `"5" + 5`
- `"5" * 3`
- `True + 1`
- `None == False`

```python
# 1. "5" + 5    -> Tahminim: Hata (TypeError) verir.
# Çünkü Python'da bir string (metin) ile bir integer (tam sayı) toplanamaz.
try:
    print("5" + 5)
except TypeError:
    print("'5' + 5 -> Hata: Metin ve sayı toplanamaz!")

# 2. "5" * 3    -> Tahminim: "555" (String çarpımı)
# Python'da bir string'i bir sayıyla çarpmak, o string'i sayı kadar yan yana yazar.
print("5" * 3)  # Çıktı: 555

# 3. True + 1   -> Tahminim: 2
# Python'da Boolean (True/False) değerleri arka planda 1 ve 0 olarak tutulur.
# True = 1 olduğu için 1 + 1 işlemi gerçekleşir.
print(True + 1) # Çıktı: 2

# 4. None == False -> Tahminim: False
print(None == False) # Çıktı: False
```

### 🔸 Soru 7:
Bir yılın artık yıl olup olmadığını kontrol edin.
Artık yıl kuralı:
- 4'e tam bölünebiliyorsa artık yıldır
- Ama 100'e bölünebiliyorsa artık yıl **değildir**
- Ama 400'e bölünebiliyorsa yine artık yıldır

```python
def artik_yil_mi(yil):
    if yil % 400 == 0:
        return True
    elif yil % 100 == 0:
        return False
    elif yil % 4 == 0:
        return True
    else:
        return False

test_yillari = [2000, 1900, 2024, 2023]

for y in test_yillari:
    sonuc = "artık" if artik_yil_mi(y) else "değil"
    print(f"{y}: {sonuc}")
```

### 🔸 Soru 8:
Bir mağazada indirim hesabı yapın:
- 500 TL altı → indirim yok
- 500-1000 TL arası → %10 indirim
- 1000-2000 TL arası → %20 indirim
- 2000 TL üstü → %30 indirim

Ödenmesi gereken tutarı yazdırın.

```python
# Cevabınızı buraya yazın:
fiyat = 1500

if fiyat < 500:
    indirim = 0
elif 500 <= fiyat < 1000:
    indirim = 0.10
elif 1000 <= fiyat <= 2000:
    indirim = 0.20
else:
    indirim = 0.30

odenecek_tutar = fiyat * (1 - indirim)
print(f"Ödenmesi Gereken Tutar: {odenecek_tutar} TL")
```

### 🔸 Soru 9:
`break` ve `continue` kullanımını gösterin:
- 1-20 arasında döngü kurun
- 3'ün katlarını `continue` ile atlayın
- 15'e ulaşınca `break` ile çıkın
- Yazdırılan sayıları gösterin

```python
print("Sayılar:")

for sayi in range(1, 21):
    if sayi == 15:
        print("15'e ulaşıldı, döngü sonlandırılıyor.")
        break

    if sayi % 3 == 0:
        continue

    print(sayi, end=" ")
```

### 🔸 Soru 10:
`while` döngüsü ile basamak sayısı bulma: Verilen bir sayının kaç basamaklı olduğunu, sayıyı 10'a bölerek bulun (string dönüşümü kullanmadan).

```python
sayi = 4752
temp = sayi
basamak_sayisi = 0

if temp == 0:
    basamak_sayisi = 1
else:
    while temp > 0:
        temp //= 10
        basamak_sayisi += 1

print(f"{sayi} -> {basamak_sayisi} basamaklı")
```

### 🔸 Soru 12:
Bir öğrenci listesi var. Her öğrenci `(isim, not)` tuple'ı olarak saklanıyor.
- Notu 70'in üzerinde olanları filtreleyin
- En yüksek notu alan öğrenciyi bulun
- Notların ortalamasını hesaplayın

```python
# Cevabınızı buraya yazın:
ogrenciler = [
    ("Ahmet", 85),
    ("Büşra", 62),
    ("Can", 91),
    ("Deniz", 74),
    ("Elif", 58),
    ("Furkan", 88)
]
gecenler = [ogrenci for ogrenci in ogrenciler if ogrenci[1] > 70]

en_yuksek = max(ogrenciler, key=lambda x: x[1])

notlar = [ogrenci[1] for ogrenci in ogrenciler]
ortalama = sum(notlar) / len(notlar)

print(f"70 Üzeri Alanlar: {gecenler}")
print(f"En Yüksek Notu Alan: {en_yuksek[0]} ({en_yuksek[1]})")
print(f"Sınıf Ortalaması: {ortalama:.2f}")
```

### 🔸 Soru 13 (Set):
İki sınıfın öğrenci listesi var. `set` kullanarak:
- Her iki sınıfta da bulunan öğrencileri bulun (kesişim)
- Toplam kaç farklı öğrenci olduğunu bulun (birleşim)
- Sadece A sınıfında olup B'de olmayan öğrencileri bulun (fark)

```python
# Cevabınızı buraya yazın:
sinif_a = {"Ali", "Ayşe", "Can", "Dilan", "Ece"}
sinif_b = {"Can", "Dilan", "Fatih", "Gökçe", "Hakan"}
kesisim = sinif_a.intersection(sinif_b)

birlesim = sinif_a.union(sinif_b) # veya sinif_a | sinif_b
toplam_ogrenci_sayisi = len(birlesim)

sadece_a = sinif_a.difference(sinif_b) # veya sinif_a - sinif_b

print(f"Her iki sınıftaki ortak öğrenciler: {kesisim}")
print(f"Toplam farklı öğrenci sayısı: {toplam_ogrenci_sayisi}")
print(f"Sadece A sınıfında olanlar: {sadece_a}")
```

### 🔸 Soru 15:
Özyinelemeli (recursive) bir fonksiyon yazın: `faktoriyel(n)` - n! hesaplasın.
- 0! = 1
- n! = n × (n-1)!

5!, 10! değerlerini hesaplayın.

```python
def faktoriyel(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktoriyel(n - 1)

sonuc_5 = faktoriyel(5)
sonuc_10 = faktoriyel(10)

print(f"5! = {sonuc_5}")
print(f"10! = {sonuc_10}")
```

### 🔸 Soru 16:
`lambda` fonksiyonu kullanarak:
- Bir sayının karesini hesaplayan lambda
- İki sayının büyüğünü döndüren lambda
- `sorted()` ile bir isim listesini son harfe göre sıralayan lambda

```python
# Cevabınızı buraya yazın:
isimler = ["Zeynep", "Ali", "Mahmut", "Su", "Berk"]
kare_al = lambda x: x**2

buyuk_olani_sec = lambda a, b: a if a > b else b

son_harfe_gore = sorted(isimler, key=lambda isim: isim[-1])

print(f"8'in karesi: {kare_al(8)}")
print(f"15 ve 42 arasından büyük olan: {buyuk_olani_sec(15, 42)}")
print(f"Son harfe göre sıralı liste: {son_harfe_gore}")
```

### 🔸 Soru 18 (Bütünleştirici):
Basit bir **öğrenci not sistemi** yazın:

- `ogrenciler` adında boş bir sözlük oluşturun (anahtar: isim, değer: notlar listesi)

- `ortalama_hesapla(isim)`: Öğrencinin not ortalamasını döndürür

```python
# Cevabınızı buraya yazın:
ogrenciler = {}

def ogrenci_ekle(isim):
    if isim not in ogrenciler:
        ogrenciler[isim] = []
        print(f"{isim} sisteme eklendi.")
    else:
        print(f"Hata: {isim} zaten sistemde kayıtlı.")

def not_ekle(isim, not_):
    if isim in ogrenciler:
        ogrenciler[isim].append(not_)
        print(f"{isim} isimli öğrenciye {not_} notu eklendi.")
    else:
        print(f"Hata: {isim} isminde bir öğrenci bulunamadı!")

def ortalama_hesapla(isim):
    if isim in ogrenciler and ogrenciler[isim]:
        notlar = ogrenciler[isim]
        return sum(notlar) / len(notlar)
    return 0

def sinif_raporu():
    print("\n--- SINIF RAPORU ---")
    toplam_ortalamalar = 0

    if not ogrenciler:
        print("Sistemde kayıtlı öğrenci yok.")
        return

    for isim in ogrenciler:
        ort = ortalama_hesapla(isim)
        toplam_ortalamalar += ort
        print(f"Öğrenci: {isim:10} | Notlar: {str(ogrenciler[isim]):15} | Ortalama: {ort:.2f}")

    genel_ortalama = toplam_ortalamalar / len(ogrenciler)
    print("-" * 30)
    print(f"Sınıf Genel Ortalaması: {genel_ortalama:.2f}\n")
# Test kodu:
# ogrenci_ekle("Ahmet")
# ogrenci_ekle("Büşra")
# ogrenci_ekle("Can")
# not_ekle("Ahmet", 85)
# not_ekle("Ahmet", 90)
# not_ekle("Ahmet", 78)
# not_ekle("Büşra", 92)
# not_ekle("Büşra", 88)
# not_ekle("Büşra", 95)
# not_ekle("Can", 70)
# not_ekle("Can", 65)
# not_ekle("Can", 73)
# sinif_raporu()
```
